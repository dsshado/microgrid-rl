```python
import os
from IPython.display import display, Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image as PILImage
import glob

FIG_DIR = os.path.join('Results', 'plots', 'paper_figures')

def show_pdf_as_png(pdf_path, dpi=150):
    """Convert PDF to PNG and display inline."""
    try:
        import fitz  # PyMuPDF
        doc = fitz.open(pdf_path)
        page = doc[0]
        mat = fitz.Matrix(dpi / 72, dpi / 72)
        pix = page.get_pixmap(matrix=mat)
        img_path = pdf_path.replace('.pdf', '_preview.png')
        pix.save(img_path)
        doc.close()
        return img_path
    except ImportError:
        return None

def show_figure(filename, title=''):
    path = os.path.join(FIG_DIR, filename)
    if not os.path.exists(path):
        print(f'Not found: {path}')
        return
    print(f'--- {title or filename} ---')
    if filename.endswith('.pdf'):
        png = show_pdf_as_png(path)
        if png:
            display(Image(filename=png))
        else:
            print('Install PyMuPDF to preview PDFs: pip install pymupdf')
            print(f'File: {path}')
    else:
        display(Image(filename=path))

# List all available figures
figs = sorted(glob.glob(os.path.join(FIG_DIR, '*')))
print('Available figures:')
for f in figs:
    print(' ', os.path.basename(f))
```

```python
# Network topology — Invalid scenario (PPO failed)
show_figure('fig_network_topology_invalid.pdf', 'Network Topology — PPO-Failed Scenario')
```

```python
# Network topology — Critical scenario
show_figure('fig_network_topology_critical.pdf', 'Network Topology — Critical Outage')
```
